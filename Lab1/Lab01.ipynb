#### Zadanie 4

Wyświetl wykres uwzględniając 10 województw (bez sumarycznego dla całego kraju), w których wydatki budżetów były największe.

> Podpowiedź: Możesz w tym celu wykorzystać moduł matplotlib.pyplot. Przykład:
```python

```

> Podpowiedź 2: Sprawdź inne funkcje aby utworzyć tytuł wykresu, oznaczyć osie, wyświetlić legendę i zmienić kolory słupków. https://matplotlib.org/stable/tutorials/introductory/pyplot.html

````python
# TUTAJ UMIESC SWOJ KOD
#filtracja
def f(row):
    return row[1] != 'POLSKA'
l2= list(filter(f,l))
l3=[]
for row in l2:
    l3.append(row[0:4] + [float(row[4])] + row[5:])
l4 = sorted(l3, key= lambda row: row[4], reverse=True)
#l4.reverse()
l5 = l4[:10]
#Values
lwoj = []
lwyd = []
for row in l5:
    lwoj.append(row[1])
    lwyd.append(row[4])
print(lwoj)
    
import matplotlib.pyplot as plt
plt.title('Wykres')
plt.xlabel('Województwa')
plt.ylabel('Wydatki')
#Zmiana nazwy
plt.xticks(rotation=90)
plt.bar(lwoj, lwyd)
````

#### Zadanie 5

Wczytaj dane z pliku `dane-ludnosc.npz`. Plik ten zawiera zbiór danych zawierający **liczbę ludnosci**.

Oblicz wydatki z budżetu na mieszkańca dla każdego z województw (_uwzględnij dane dla roku `2008`_).


````python
# TUTAJ UMIESC SWOJ KOD
import numpy as np
with open('./dataset/dane-ludnosc.npz', 'rb') as file2:
    content = np.load(file2)
    columns, data = content['columns'].tolist(), content['data'].tolist()
print("Załadowano bazę danych")
print(columns)
#print (data)
#print(l)
l1 =l
l2= list(filter(lambda row: row[4] == '2008',data))
lfinal = []
for x in range(17):
    lfinal.append([l1[x][1],float(l1[x][4])/float(l2[x][5])])
print(lfinal)

#Values
lwoj = []
lwyd = []
for row in lfinal:
    lwoj.append(row[0])
    lwyd.append(row[1])
print(lwoj)
    
import matplotlib.pyplot as plt
plt.title('Wykres')
plt.xlabel('Województwa')
plt.ylabel('Wydatki na osobe')
#Zmiana nazwy
plt.xticks(rotation=90)
plt.bar(lwoj, lwyd)
````

#### Zadanie 6

Wykonaj dalszą analizę zbiorów danych (dowolną). Czy coś ciekawego zaobserwowałeś?


````python
# TUTAJ UMIESC SWOJ KOD
# Sprawdzmy czy średnia dla wszystkich województw jest taka sama jak wartość dla polski
avg = 0
for row in lfinal:
    if row[0]!= 'POLSKA':
        avg += row[1]
print(avg/16)
print(lfinal[0])
# Województwo mazowieckie ma znacznie większe wyfatki na mieszkańca ale również dużą liczbę ludności co znacznie podnosi
# średnią w porównaniu do liczenia jej z poszczególnych woj.
````
