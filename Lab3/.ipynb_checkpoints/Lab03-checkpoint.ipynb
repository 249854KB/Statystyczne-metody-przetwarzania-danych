## ZADANIE 

````python
## Inportowanie 

import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt 

with open('./dataset/dataset.npz', 'rb') as f:
    data = np.load(f)
    train, test = data['train'], data['test']
    
````

````python
## Por贸wnywanie 

def accuracy(classes, features, k_neighbour):
    #Jeli na kolumnie 0 jest taka klasa to okay
    train_new = train[np.isin(train[:,0], classes)] 
    test_new =  test[np.isin(test[:,0], classes)] 
    
    values = train_new[:,features] # Te cechy, kt贸re chcemy por贸wna
    labels = train_new[:,0] #labels
    
    neigh = KNeighborsClassifier(n_neighbors=k_neighbour)
    neigh.fit(values, labels) #Tworzymy specyfikator
    
    guess = []
    for testers in range(len(test_new)):
        guess.append( neigh.predict([test_new[testers,features]]))
        guess[testers] = guess[testers][0] #fixing broken values # ID OF EDLISTSTS FOR TESTER
    
   
    # guess[:] mamy wszystkie kategorie kt贸re zgadlimy 
    # test_new[:,0] mamy poprawne labelki
    
    #print(guess[:])
    #print(test_new[:,0])
    T=0
    for i in range(len(test_new)):
        if test_new[i,0] == guess[i]:
            T=T+1
        
    Accuracy = T / len(test_new) * 100
    
    print(Accuracy)
    
    cm =confusion_matrix(test_new[:,0], guess[:])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels = classes)
    disp.plot()

    plt.show()
````

````python
# Zmiana liczby ssiad贸w
klasy = [9,10]
cechy = [4,8]
sasiedzi = 1
accuracy(klasy,cechy,sasiedzi)
sasiedzi = 3
accuracy(klasy,cechy,sasiedzi)
sasiedzi = 5
accuracy(klasy,cechy,sasiedzi)
sasiedzi = 8
accuracy(klasy,cechy,sasiedzi)

# Obserwacje: Dla tego przykadu wzrost dokadnoci nie jest tak du偶y jak w innych przypadkach
````

````python
# Zwikszanie liczby klas
klasy = [9,10]
cechy = [4,8]
sasiedzi = 3
accuracy(klasy,cechy,sasiedzi)
klasy = [9,10,12,13]
accuracy(klasy,cechy,sasiedzi)
# Obserwacje - Im wicej klas tym wiksza dokadno
````

````python
# Zwikszanie liczby Cech
klasy = [9,10]
cechy = [4,8]
sasiedzi = 3
accuracy(klasy,cechy,sasiedzi)
cechy = [4,8,7]
accuracy(klasy,cechy,sasiedzi)
cechy = [4,8,7,11]
accuracy(klasy,cechy,sasiedzi)
cechy = [4,8,9,10]
accuracy(klasy,cechy,sasiedzi)
# Obserwacje: Im wicej cech tym wiksza skuteczno 
````
